```python
def read_log():
    with open('fitbit/2018-04-26_through_2018-05-26.csv') as f:
        lines = f.read()
        return lines
```

```python
def split_content(lines):
    lines = lines.split('\n')
    return lines
```

```python
def split_lines(lines):
    new_lines = []
    for line in lines:
        new_lines.append(line.split(' '))
    return new_lines
```

```python
lines = split_lines(lines)
```
